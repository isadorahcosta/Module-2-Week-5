```python
# Only interested in the pet attributes, not orders
```

```python
pets=df[['customer_id','pet_id', 'gender', 'neutered', 'pet_breed_size', 'pet_life_stage_at_order', 'pet_allergen_list',
         'pet_fav_flavour_list','pet_health_issue_list',
         'pet_has_active_subscription',
         'ate_wet_food_pre_tails', 'pet_food_tier', 'wet_food_textures_in_order'
        ]]
pets.sample(20)
```
